Importing spaCy

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

Create a Doc object from the file owlcreek.txt


```python
with open("owlcreek.txt", "r", encoding="utf-8") as f:
    text = f.read()

doc = nlp(text)
```

How many tokens are contained in the file?

```python
print(len(doc))
```

How many sentences are contained in the file?

```python
sentences = list(doc.sents)
print(len(sentences))
```

Print the second sentence in the document

```python
print(sentences[1])
```

For each token in the sentence above, print its text, POS tag, dep tag and lemma.

```python
for token in sentences[1]:
    print(token.text, token.pos_, token.dep_, token.lemma_)
```

Write a matcher called 'Swimming' that finds both occurrences of the phrase "swimming vigorously" in the text.

```python
from spacy.matcher import Matcher
doc = nlp(text)
matcher = Matcher(nlp.vocab)
pattern = [
    {"TEXT": "swim"},
    {"TEXT": "vigorously"}
]
matcher.add("Swimming", [pattern])
matches = matcher(doc)

print("Number of matches:", len(matches))
for match_id, start, end in matches:
    print("Match:", doc[start:end].text)
```

```python
for token in doc:
    if token.text.lower() == "swimming":
        print(doc[token.i:token.i+5])
```

Print the text surrounding each found match.


```python
matches = matcher(doc)
for match_id, start, end in matches:
    print(doc[start-5:end+5].text)
```
